```python
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import san
from matplotlib import style
style.use('ggplot')

from IPython.display import display

%matplotlib notebook

from san import Batch
from_date = '2018-06-01'
to_date = '2018-10-16'

batch = Batch()
batch.get(
    "daily_active_addresses/santiment",
    from_date=from_date,
    to_date=to_date,
    interval="1d"
)

batch.get(
    "prices/santiment",
    from_date=from_date,
    to_date=to_date,
    interval="1d"
)

[daa, prices] = batch.execute()
print(daa.head(5))
print(prices.head(5))
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(daa.index.values, daa[['activeAddresses']], 'g-')
ax2.plot(prices.index.values, prices[['priceUsd']], 'b-')

ax1.set_xlabel('Datetime')
ax1.set_ylabel('Daily Active Addresses', color='g')
ax2.set_ylabel('Price $', color='b')

fig.autofmt_xdate()

plt.show()
```
